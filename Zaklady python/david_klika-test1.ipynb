Vytvořte aplikaci, která bude přijímat tři vstupy uživatele.

* prvním bude výška v cm
* druhý váha v kg
* třetí věk

Vypočítejte mi BMI index pro tyto údaje. BMI = (váha v kg) děleno (výška v metrech na druhou)

Výpis napište ve formátu:

Uživatel ve věku "věk" má BMI: "BMI" což znamená Nadváha/ Podváha atd.

viz hodnoty níže.
* 0 až 18,5: Hodnota podváhy
* 18,5 až 25: Hodnota ideální váhy
* 25 až 30: Hodnota nadváhy
* 30 až 40: Hodnota obezity prvního stupně
* 40 a více: Hodnota těžké obezity

````python

vyska= float(input("Zadejte výšku v cm: "))
vaha = float(input("Zadejte váhu v kg: "))
vek = int(input("Zadejte věk: "))


vyska = vyska / 100

bmi = vaha / (vyska ** 2)


if bmi < 18.5:
    stav = "Hodnota podváhy"
elif bmi < 25:
    stav = "Hodnota ideální váhy"
elif bmi < 30:
    stav = "Hodnota nadváhy"
elif bmi < 40:
    stav = "Hodnota obezity prvního stupně"
else:
    stav = "Hodnota těžké obezity"

print(f'Uživatel ve věku "{vek}" má BMI: "{bmi}" což znamená {stav}')
````


**Refereční řešení**

```
Zadejte vaší výšku v cm.: 187.7
Zadejte vaší váhu v kg.: 55.8
Zadejte vaší věk: 35
Uživatel ve věku 35 má BMI: 15.84 což znamená Podváha
```
